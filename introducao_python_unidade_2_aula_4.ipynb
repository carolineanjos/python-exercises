```python
import matplotlib.pyplot as plt

#Dados
x = [1,2,3,4,5]
y = [2,4,1,3,5]

#Criar um gráfico de linha
plt.bar(x,y)

#Adicionar rótulos aos eixos
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

#Adicionar um título para o gráfico
plt.title("Gráfico de linha")

#Mostrar o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

#Dados de exemplo
meses = ["Janeiro", "Fevereiro", "Marco", "Abril", "Maio"]
vendas = [120, 90, 150, 70, 200]

plt.bar(meses, vendas)

plt.xlabel("Meses")
plt.ylabel("Vendas")

plt.title("Informe de vendas")

plt.show()
```
